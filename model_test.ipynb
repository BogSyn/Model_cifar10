```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt

# Завантажуємо модель
model = tf.keras.models.load_model('model_2_finetuned.h5')

# Завантажуємо та обробляємо зображення
img_path = 'test_jpg/35.jpg'  # замініть шлях до зображення
img = image.load_img(img_path, target_size=(56, 56)) # Змінюємо розмір на 56x56
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)  # Додаємо розмір для batch
img_array = img_array / 255.0  # Нормалізуємо значення пікселів

# Передбачаємо клас зображення
predictions = model.predict(img_array)
predicted_class = np.argmax(predictions[0])

# Назви класів у CIFAR-10
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Відображаємо результат
plt.imshow(img)
plt.title(f'Predicted: {class_names[predicted_class]}')
plt.show()

print(f"Model's prediction: {class_names[predicted_class]} with confidence {predictions[0][predicted_class]:.2f}")
```
